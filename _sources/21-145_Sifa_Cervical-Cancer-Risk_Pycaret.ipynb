## PREPROCESSING

### 1. Normalisasi Menggunakan Min-Max Scaler

Normalisasi Min-Max Scaler adalah salah satu metode normalisasi yang digunakan dalam pengolahan data dan pembelajaran mesin. Tujuan normalisasi adalah untuk mengubah nilai-nilai variabel dalam dataset sehingga mereka dapat memiliki skala yang seragam atau berada dalam rentang tertentu.

Normalisasi Min-Max Scaler digunakan untuk mengubah setiap nilai dalam dataset ke dalam rentang yang ditentukan, biasanya antara 0 dan 1.

Rumus dari normalisasi Min-Max Scaler yaitu sebagai berikut :

$Xnormalized = \frac{(X - Xmin)}{Xmax - Xmin}$

* Xnormalized adalah nilai yang telah dinormalisasi
* X adalah nilai asli dari variabel
* Xmin adalah nilai minimum dari variabel
* Xmax adalah nilai maksimum dari variabel

Dengan menggunakan rumus ini, setiap nilai dalam variabel akan diubah sedemikian rupa sehingga berada dalam rentang antara 0 dan 1. Proses ini membantu untuk mengatasi perbedaan skala antara variabel-variabel dalam dataset, sehingga variabel-variabel tersebut dapat memberikan kontribusi seimbang dalam analisis atau model pembelajaran mesin.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from pickle import dump
```

### 2. Melakukan Standarisasi pada Data Train

```python
#Membaca data dari file csv
dt = pd.read_csv('https://raw.githubusercontent.com/niswatulsifa/PSD/main/cervical-cancer.csv')

#Memisahkan fitur (X) dan label (y)
X = dt.drop(['ca_cervix'], axis = 1) #Menghapus kolom ca_cervix karena pada variabel X hanya berisi fitur saja
y = dt['ca_cervix'] #Inisialisasi kolom label ca_cervix

#Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1, test_size = 0.2)
```

```python
#Mendefinisikan Scaler
minmax = MinMaxScaler()
#Menghitung rata-rata dan standart deviasi tiap fitur (menggunakan rumus Min-Max Scaler)
minmax.fit(X_train)
#Mentransformasi nilai-nilai dari setiap fitur yang dihitung sebelumnya. Hasilnya adalah data latih dan data uji yang telah di skalakan.
X_train_scaled = minmax.transform(X_train)
X_test_scaled = minmax.transform(X_test)
#Menyimpan model yang telah dilatih
dump(minmax, open('cervical-cancer.pkl','wb'))
```

### 3. Reduksi Dimensi menggunakan PCA

Principal Component Analysis (PCA) adalah suatu metode statistik yang digunakan untuk mengidentifikasi pola dalam data multivariat dengan mereduksi dimensi variabel. PCA mencoba untuk mengekspresikan data dalam bentuk kombinasi linear dari variabel-variabel yang ada sehingga menjadi dimensi baru, yang disebut principal components dimana menggambarkan variasi data dengan baik. Berikut adalah rumus PCA dan penjelasannya:

1. Hitung Matriks Kovarian:

   $ \text{Cov}(\mathbf{X}) = \frac{1}{n-1} \sum_{i=1}^{n} (\mathbf{X}_i - \bar{\mathbf{X}})(\mathbf{X}_i - \bar{\mathbf{X}})^T $

    di mana:


  *   $\mathbf{X}_i \text{ adalah vektor data poin ke-i,}$
  *   $\bar{\mathbf{X}} \text{ adalah vektor rata-rata dari semua data,}$
  *   $n \text{ adalah jumlah data poin.} $

2. Hitung Nilai dan Vektor Eigen: Menghitung nilai eigen (λ) dan vektor eigen (v) dari matriks kovarian.

   $ \text{Cov}(\mathbf{X})\mathbf{v} = \lambda\mathbf{v} $

3. Pilih Principal Components: Principal components dipilih berdasarkan nilai eigen tertinggi. Misalnya, jika kita ingin mereduksi dimensi menjadi k, kita memilih k vektor eigen dengan nilai eigen tertinggi.

4. Transformasi Data: Data asli (X) dapat diproyeksikan ke dalam ruang yang baru (Y) yang terdiri dari principal components yang dipilih.

   $ \mathbf{Y} = \mathbf{X} \mathbf{V} $

    di mana:

*   (Y) : matriks yang berisi data yang telah diproyeksikan ke dalam ruang principal components,
*   (X) : matriks data asli,
*   (V) : matriks yang berisi vektor eigen sebagai kolom-kolomnya.

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=15)
pca_train = pca.fit_transform(X_train_scaled)
```

pca = PCA(n_components=15) yaitu membuat objek PCA dengan parameter n_components yang ditetapkan yaitu 15, yang berarti akan menghasilkan 15 komponen utama (principal components) dari data. Ini adalah jumlah dimensi yang diinginkan setelah PCA.

pca_train = pca.fit_transform(X_train_scaled) yang dimana menggunakan objek PCA yang telah dibuat untuk melakukan PCA pada data training yang telah di-scaling (X_train_scaled). Fungsi fit_transform digunakan untuk menghitung komponen utama dan secara efektif mentransformasikan data ke dalam ruang dimensi yang lebih rendah.

```python
#Digunakan untuk mengetahui jumlah baris dan kolom dari data hasil PCA pada data uji
pca_train.shape
```

```python
#Digunakan untuk mentransformasi data uji menjadi fitur baru yang dibentuk oleh komponen-komponen PCA
pca_test = pca.transform(X_test_scaled)
```

```python
#Digunakan untuk mengetahui tentang jumlah baris dan kolom dari data hasil transformasi PCA pada data latih.
pca_test.shape
```

## KLASIFIKASI MENGGUNAKAN RANDOM FOREST CLASSIFIER

Random Forest merupakan salah satu metode ensemble learning yang digunakan dalam machine learning untuk meningkatkan kinerja model. Ensemble learning menggabungkan beberapa model pembelajaran untuk menciptakan model yang lebih kuat dan lebih stabil. Random Forest khususnya digunakan untuk tugas klasifikasi dan regresi.

**Bagaimana Random Forest Bekerja:**

1. Pemilihan Acak Sampel (Random Sampling):
   - Dari dataset pelatihan, dibuat sejumlah dataset sampel yang sama dengan penggantian (bootstrap samples).
   - Bootstrap samples adalah dataset acak yang dibuat dengan mengambil sampel secara acak dengan penggantian dari dataset pelatihan.

2. Pembangunan Pohon Keputusan (Decision Trees):
   - Untuk setiap bootstrap sample, sebuah pohon keputusan (decision tree) dibangun.
   - Pohon keputusan dibangun dengan membagi dataset pada setiap simpul berdasarkan fitur-fitur tertentu.

3. Voting (Voting Scheme):
   - Setiap pohon memberikan prediksi untuk setiap instance yang masuk.
   - Prediksi dari setiap pohon dijumlahkan dan mayoritas diambil sebagai prediksi akhir.

**Rumus Random Forest:**

1. Random Sampling:
   - Untuk setiap pohon keputusan \( k \), dataset acak \( D_k \) dibuat dengan penggantian.
   - Ukuran dataset acak sama dengan ukuran dataset pelatihan, tetapi beberapa instance dapat muncul beberapa kali dan yang lainnya mungkin tidak muncul sama sekali.

2. Pembangunan Pohon Keputusan:
   - Pohon keputusan dibangun berdasarkan pemilihan fitur yang acak.
   - Sebuah simpul membagi dataset berdasarkan fitur tertentu yang dipilih secara acak.
   - Proses ini diulang hingga terbentuk pohon keputusan lengkap atau memenuhi kriteria berhenti tertentu.

3. Voting:
   - Prediksi dari setiap pohon dijumlahkan dan mayoritas diambil sebagai prediksi akhir.
   - Misalnya, jika kita memiliki \( N \) pohon dan \( n \) pohon memprediksi kelas A dan \( m \) pohon memprediksi kelas B, dengan \( n > m \), maka prediksi akhir akan menjadi kelas A.

**Keuntungan Random Forest:**
- Mengatasi overfitting karena menggunakan beberapa pohon yang bekerja secara bersamaan.
- Robust terhadap outliers dan noise dalam data.
- Mampu menangani dataset besar dengan fitur yang banyak.

**Rumus Tambahannya untuk Regresi:**
Jika Random Forest digunakan untuk tugas regresi, prediksi akhir dapat dihitung sebagai rata-rata prediksi dari semua pohon.

$[ \text{Prediksi akhir} = \frac{1}{N} \sum_{i=1}^{N} \text{Prediksi pohon ke-i} ]$

Di mana \( N \) adalah jumlah pohon dalam Random Forest.

**Code Random Forest Classifier**
